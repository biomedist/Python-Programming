**i부터 n까지의 합(3단 전략)**

```python
i = 1
sum = 0
n = int(input("합산할 숫자를 입력하세요:"))
while i <= n :
    sum = sum + i
    i += 1

print("1부터", str(n)+"까지의 합은", str(sum)+"입니다.")
```

# 6강 2절

**반지름은 10이고 높이가 1, 5, 14, 26, 31인 원뿔의 부피와 겉넓이를 각각 출력하는 문제**

```python
rad = 10
hei = 25
# 리스트 생성
hei_list = [1, 5, 14, 26, 31]  #리스트 생성후 hei_list변수에 넣음

for hei in hei_list :
    # 모선 계산
    sla = pow(rad**2 + hei ** 2, 0.5)
    # 부피 & 겉넓이 출력
    vol = 1/3 * 3.14 * rad ** 2 * hei
    suf = 3.14 * rad ** 2 + 3.14 * rad * sla
    print("반지름", str(rad)+"높이", hei, "원뿔")
    print("부피", vol, '겉넓이:', suf)

```

**유클리드 호제법**

```python
# 유클리드 호제법으로 두 수의 최대공약수(GCD)를 구하는 프로그램

# 두 수 입력 받기
a = int(input("첫 번째 숫자를 입력하세요: "))
b = int(input("두 번째 숫자를 입력하세요: "))

# a를 b로 나눈 나머지를 b에 저장
# b가 0이 될 때 까지 반복

while b != 0 :      # b가 0이 아니면 계속 반복
    rem = a % b
    a = b
    b = rem
print("두 수의 최대공약수는", str(a)+"입니다.") #최대공약수가 왜 a일까
# 유클리스 호제법의 핵심: 두 수의 최대공약수는 (큰 수, 작은 수)의 GCD와 (작은 수, 나머지)의 GCD가 같다
# GCD(a, b) = GCD(b, a % b)
```
