## 원뿔의 부피&겉넓이 계산 프로그램 5


```python
rad = input()
print(type(rad))
print(rad)
```

```python
# 원뿔 부피 및 겉넓이 계산 프로그램
# input함수는 모든것을 문자로 받아들이므로 숫자로 변환하는 작업이 필요하다
rad = int(input("반지름을 입력하세요"))
hei = int(input("높이를 입력하세요"))
sla = pow(rad ** 2 + hei ** 2, 0.5)
#부피 & 겉넓이 출력
print(1 / 3 * 3.14 * rad ** 2 * hei )
print(3.14 * rad ** 2 + 3.14 * rad * sla)
```
